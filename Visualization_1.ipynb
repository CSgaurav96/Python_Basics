```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import datetime
```

* Matplotlib is a library in python & 'pyplot' is the module in the library  
* matplotlib.pyplot is a collection of functions that make Matplotlib work like MATLAB. It provides a convenient way to create various types of plots and visualizations

## Plotting a single Line 

```python
# X-Y Plot
plt.plot([1,2,3,4], [16,4,1,8])
plt.show()
```

* plt.plot is a function in the matplotlib.pyplot module that is used to create a variety of plots, with the most common being a line plot.

```python
plt.plot([10,20,30,40])
plt.show()
```

* we can add more parameters to the 'plot' function

```python
plt.plot([1,2,3,4], [16,4,1,8], marker='o', linestyle='-', color='b', label='Data')
plt.show()
# Here we can see that label is not showing 
```

## Plotting multiple lines 

```python
# We have 4 lines or 4 numpy arrays
t = np.arange(0,2.5,0.1)
y1 = np.sin(math.pi*t)
y2 = np.sin(math.pi*t + math.pi/2)
y3 = np.sin(math.pi*t - math.pi/2)
```

* Here array values are appeared on the y-axis
* The index of this array is used as the x-axis : 0,1,2,3

```python
# First array ('t') values would show up on the x-axis and remaining arrays values would show up on the y-axis
plt.plot(t,y1,t,y2,t,y3)
```

```python
plt.plot(t,y1)
plt.plot(t,y2)
plt.plot(t,y3)
```

* Here we can notice that it created all the plots in the same graph. 
* In Matplotlib, each call to plt.plot adds a new line to the current plot.
*  If you want to create a new graph (new figure) after plotting one set of data and then start fresh with another set of data, you can use plt.figure() to create a new figure.

```python
plt.plot(t,y1)
plt.figure()

plt.plot(t,y2)
plt.plot(t,y3)
```

* Another way of creating multiple line graphs 

## Subplots 

* The subplot() function takes 3 arguments that describes the layout of the figure
* The layout is organised in rows and columns, which are represented by the first and the second argument
* The 3rd argument represents the index of the current plot

### Create 2 plots on the top of each others

```python
x = np.array([0,1,2,3])
y1 = np.array([3,8,1,10])
y2 = np.array([10,20,30,40])

plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)


plt.show()
```

## Formatting of axis 

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis([0,5,0,20])
plt.show()
```

* This code is setting the axis limits for the plot
* The axis function takes a list of four numbers as input, which represent the minimum and maximum values for the x-axis and y-axis, respectively
* Any data points outside of this range will not be shown on the plot.

```python
# We set the title, Xlabel, Ylabel

plt.title('My First Plot')
plt.xlabel('Counting')
plt.ylabel('Square Values')
plt.axis([0,5,0,20])
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
```

```python
plt.xlabel('Counting', color = 'green')
plt.ylabel('Square Values', color = 'green')
plt.axis([0,5,0,20])
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.sin(3*x)/x
plt.xticks([-2*np.pi ,-np.pi , 0, np.pi , 2*np.pi] , [r'$-2\pi$' , r'$-\pi$', r'$0$' , r'$\pi$' , r'$2\pi$'])
plt.xticks(rotation = 45) # It change the alignment of the xticks labels
plt.yticks([-1,0,1,2,3], [r'$-1$', r'$0$' , r'$+1$' ,r'$+2$',r'$+3$'])
plt.plot(x,y)
```

## Adding labels within the graph space 

```python
# Here we are adding labels within the graph area based on the (x,y) coordinates
plt.axis([0,5,0,20])
plt.text(1,1.5,'first')
plt.text(2,4.5,'second')
plt.text(3,9.5, 'third')
plt.text(4,16.5,'forth')
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

```python
plt.axis([0,5,0,20])
plt.text(1,1.5,'first')
plt.text(2,4.5,'second')
plt.text(3,9.5, 'third')
plt.text(4,16.5,'forth')
plt.text(1.1,12,r'$y = x^2$', fontsize = 20 , bbox = {'facecolor' : 'yellow', 'alpha':0.9})
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

* Here 'alpha' is a parameter that sets the transparency level of the background color of the text box
* value of alpha ranges from 0 (completely transparent) to 1 (completely opaque)
* The $...$ around y = x^2 is a notation used in LaTeX to indicate that the content within the dollar signs should be rendered as mathematical notation.
* The r before the string indicates that it is a raw string, which means that any special characters within the string will be interpreted as literal characters, rather than as special commands.

```python
# Gridlines added
plt.axis([0,5,0,20])
plt.text(1,1.5,'first')
plt.text(2,4.5,'second')
plt.text(3,9.5, 'third')
plt.text(4,16.5,'forth')
plt.text(1.1,12,r'$y = x^2$', fontsize = 20 , bbox = {'facecolor' : 'yellow', 'alpha':0.9})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
```

## Legends

```python
plt.axis([0,5,0,20])
plt.text(1,1.5,'first')
plt.text(2,4.5,'second')
plt.text(3,9.5, 'third')
plt.text(4,16.5,'forth')
plt.text(1.1,12,r'$y = x^2$', fontsize = 20 , bbox = {'facecolor' : 'yellow', 'alpha':0.9})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.legend(['First Series'])
```

* The legend code has to come after the graph plot command
*  The reason it is being placed at the end of the code is so that it is displayed on top of the other elements of the plot, rather than being obscured by them. Additionally, this ensures that all the elements of the plot have been created before the legend is added, allowing it to reference all the necessary elements.

```python
plt.axis([0,5,0,20])
plt.text(1,1.5,'first')
plt.text(2,4.5,'second')
plt.text(3,9.5, 'third')
plt.text(4,16.5,'forth')
plt.text(1.1,12,r'$y = x^2$', fontsize = 20 , bbox = {'facecolor' : 'yellow', 'alpha':0.9})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.plot([1,2,3,4],[0.8,3.5,8,15],'g^')
plt.plot([1,2,3,4],[0.5,2.5,4,12],'b*')
plt.legend(['First Series','Second Series', 'Third Series'] , loc = 2)
```

* The loc parameter is used to specify the location of the legend
* loc=1: upper right
* loc=2, upper left
* loc=3, lower left
* loc=4, lower right

```python
plt.axis([0,5,0,20])
plt.text(1,1.5,'first')
plt.text(2,4.5,'second')
plt.text(3,9.5, 'third')
plt.text(4,16.5,'forth')
plt.text(1.1,12,r'$y = x^2$', fontsize = 20 , bbox = {'facecolor' : 'yellow', 'alpha':0.9})
plt.grid(True)
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.plot([1,2,3,4],[0.8,3.5,8,15],'g^')
plt.plot([1,2,3,4],[0.5,2.5,4,12],'b*')
plt.legend(['First Series','Second Series', 'Third Series'] , bbox_to_anchor=(1.05, 1), loc=2)
```
