# Subplot function
* The function takes 3 arguments which determines the layout of the figure
* layout is organized in rows and columns, which are represented by the first and second argument.
* The third argument represents the index of the current plot.
* E.g. plt.subplot(1, 2, 1) : the figure has 1 row, 2 columns, and this plot is the first plot.
* E.g. plt.subplot(1, 2, 2) : the figure has 1 row, 2 columns, and this plot is the 2nd plot.

### Subplots arranged in horizontal order

```python
# plot 1 
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1) # assignment of the plot
plt.plot(x,y)

# plot 2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

### Subplots arranged in vertical order

```python
# plot 1 
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,1,1) # assignment of the plot
plt.plot(x,y)

# plot 2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(2,1,2)
plt.plot(x,y)

plt.show()
```

### Alternate way to create subplots 

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
```

```python
fig , ax = plt.subplots(1,2 , figsize = (15,5))
ax[0].plot(x,y)   # Here ax[0] is used because we have only one row in the diagram
ax[1].plot(x,y)
plt.show()
```

```python
fig , ax = plt.subplots(2,2 , figsize = (12,5))
ax[0,0].plot(x,y) # Here we used a list ax[0,0] because we have multiple rows in the diagram
ax[0,1].plot(x,y)
ax[1,0].plot(x,y)
ax[1,1].plot(x,y)
plt.show()
```

* Here we didn't had to use 'plt' before creating the plot. 

# Contour Plots

* Sometimes it is useful to display three-dimensional data in two dimensions using contours or color-coded regions.
* There are three Matplotlib functions that can be helpful for this task: plt.contour for contour plots, plt.contourf for filled contour plots, and plt.imshow for showing images. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

### Visualization of a 3 dimentional function 

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0,5,10)
y = np.linspace(0,5,10)
```

```python
X, Y = np.meshgrid(x,y)
```

```python
Z = f(X,Y)
```

```python
plt.contour(X, Y, Z)
```

* by default when a single color is used, negative values are represented by dashed lines, and positive values by solid lines.
* the lines can be color-coded by specifying a colormap with the cmap argument.

```python
# It fills the area with the same color which have the same Z value 
plt.contourf(X, Y, Z , alpha = 0.5, colors = 'red');
```

* Here 'alpha' controls the brightness of the filled areas. 
* 'alpha' lies between 0 (complelely transparent) and 1 (solid filled)
