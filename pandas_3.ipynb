* By default, pd.read_csv() assumes the delimiter is a comma (,)
* pd.read_table() assumes the delimiter is a tab (\t)
* A delimiter is a character or sequence of characters that separates values in a text file.It is used to distinguish one piece of data from another. The most common delimiters are commas (,) and tabs (\t).
* A tab delimiter (\t) is a whitespace character that is used to separate values in a text file. It is often used in data files where values in a row are separated by a consistent amount of whitespace, typically a single tab character. The tab character is represented by the escape sequence \t.
