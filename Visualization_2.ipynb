## Time-Series graphs 

```python
import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
```

```python
events= [datetime.date(2015,1,23),    # date format : year/month/day
        datetime.date(2015,1,28),
        datetime.date(2015,2,3),
        datetime.date(2015,2,21),
        datetime.date(2015,3,15),
        datetime.date(2015,3,24),
        datetime.date(2015,4,8),
        datetime.date(2015,4,24)]
reading = [12, 22, 25, 20, 18, 15, 17, 14]
plt.plot(events, reading)
```

```python
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%Y - %m')

events= [datetime.date(2015,1,23),
        datetime.date(2015,1,28),
        datetime.date(2015,2,3),
        datetime.date(2015,2,21),
        datetime.date(2015,3,15),
        datetime.date(2015,3,24),
        datetime.date(2015,4,8),
        datetime.date(2015,4,24)]
reading = [12, 22, 25, 20, 18, 15, 17, 14]
fig, ax = plt.subplots()
plt.plot(events, reading)

ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(timeFmt)
ax.xaxis.set_minor_locator(days)
```

### Plotting data using data frame 

```python
data = {'series_1': [1,3,4,3,5], 'series_2': [2,4,5,2,4], 'series_3': [3,2,3,1,3]}
df = pd.DataFrame(data)
x = np.arange(5)
plt.axis([0,5,0,7])
plt.plot(x,df)
plt.show()
```

### Plotting data using data frame : Legends

```python
data = {'series_1': [1,3,4,3,5], 'series_2': [2,4,5,2,4], 'series_3': [3,2,3,1,3]}
df = pd.DataFrame(data)
x = np.arange(5)
plt.axis([0,5,0,7])
plt.plot(x,df)
plt.legend(data, loc =2) 
plt.show()
```

* Here we can notice that it takes the column names of the dataframe as legend names

## Scattar Plot 

```python
x = [1,2,3,4]
y = [2,4,1,3]
z = [0,1,0,1]
```

**Creating scatterplot using the libarary 'pyplot'**

* We will use the 'scatter' function and pass in the 'x' and 'y' data for the points, along with any desired formatting options (color, marker & size)

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y , label = "Train") # These labels shows up when we use legend()
plt.scatter(x,z , label = "Test")
plt.legend()
plt.show()
```

**Creating scatterplot using the libarary 'Seaborn'**

```python
import seaborn as sns
```

```python
sns.scatterplot(x,y)
```

```python
sns.scatterplot(x,y , hue = z)
```

```python
# s: size of dots
sns.scatterplot(x,y , hue = z , s = 100)
```

```python
# 'alpha' determines the darkness of various dots
sns.scatterplot(x,y , hue = z , s = 100 , alpha = 0.5)
```

### Pie Chart 
* plot.pie is the function, using that we will create a pie chart. 

```python
import pandas as pd
import matplotlib.pyplot as plt
data_path = "C:\Personal\Carrier Path\Data_Scientist\Accumulated_content\AI\Implementation_project_pro\Various_Projects\Final Projects\Credit Card Default prediction\input/"
df = pd.read_csv(data_path + "cs-training.csv")
df.head()
```

```python
fig , ax = plt.subplots(figsize = (7,7))
df['SeriousDlqin2yrs'].value_counts().plot.pie(explode = [0, 0.1] , autopct = '%1.1f%%')
plt.show()
```

**explode**
* This parameter is used to explode (separate) the second slice of the pie chart (corresponding to the value '1' in 'SeriousDlqin2yrs') to highlight it.
***
**autopct = '%1.11f%%'**
* This parameter specifies the format of the autopct (automatic percentage) labels on the pie chart, displaying one decimal place.
